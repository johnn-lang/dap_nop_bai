```python
import pandas as pd
```

```python
data=pd.read_csv('https://github.com/lenhattung/SP26-AI1904-DAP391m/raw/refs/heads/main/lab_part_2/01%20-%20Linear%20and%20logistic%20regression/titanic.csv')
```

```python
data=pd.get_dummies(data, columns=['sex', 'embarked','pclass','boat','cabin'])
data.head(5)
```

```python
data
```

```python
df = data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df.drop(columns=['passenger_id','name'], inplace=True)
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

k_fold = 5
fold_size = len(data) // k_fold

X = df.drop(columns=['survived','ticket','home.dest'])
y = df['survived']

for k in range(k_fold):
    start = k * fold_size
    end = (k + 1) * fold_size

    X_test = X.iloc[start:end]
    y_test = y.iloc[start:end]

    X_train = pd.concat([X.iloc[:start], X.iloc[end:]])
    y_train = pd.concat([y.iloc[:start], y.iloc[end:]])

    model=LinearRegression()
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    r2=r2_score(y_test, y_pred)
    print(f"Fold {k + 1}: R-squared = {r2} ,  RMSE = {np.sqrt(mean_squared_error(y_test, y_pred))}")

```

```python
data_white=pd.read_csv('winequality-white.csv')
data_red=pd.read_csv('winequality-red.csv')
```

```python
data_white = data_white.copy()
data_white["wine_type"] = "white"

data_red = data_red.copy()
data_red["wine_type"] = "red"

data_wine = pd.concat([data_white, data_red], ignore_index=True)
```

```python
mixed_col = (
    'fixed acidity;"volatile acidity";"citric acid";"residual sugar";'
    '"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";'
    '"pH";"sulphates";"alcohol";"quality"'
)
new_cols = [
    "fixed_acidity", "volatile_acidity", "citric_acid", "residual_sugar",
    "chlorides", "free_sulfur_dioxide", "total_sulfur_dioxide", "density",
    "pH", "sulphates", "alcohol", "quality"
]

df_split = data_wine[mixed_col].str.split(";", expand=True)
df_split.columns = new_cols
df_split = df_split.astype(float)

data_wine = pd.concat(
    [data_wine.drop(columns=[mixed_col]), df_split],
    axis=1
)

```

```python
import statsmodels.api as sm

formula = """
quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar
+ chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density
+ pH + sulphates + alcohol + C(wine_type)
"""

mod = sm.GLM.from_formula(
    formula=formula,
    data=data_wine
)

res = mod.fit()
res.summary()
```
